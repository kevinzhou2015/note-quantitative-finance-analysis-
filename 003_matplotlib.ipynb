```python
#简介
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontManager, FontProperties
import numpy as np
import matplotlib.dates as mdate
```

```python
ChinaBank = pd.read_csv('datasets/ChinaBank.csv',index_col = 'Date')
ChinaBank.index = pd.to_datetime(ChinaBank.index)
ChinaBank = ChinaBank.iloc[:,1:]
Close = ChinaBank['Close']
Open=ChinaBank.Open
plt.plot(Close['2014'])
plt.show()
```

```python
#2修改绘图属性
#2.1 坐标
plt.plot([1,1,0,0,-1,0,1,1,-1])
plt.ylim(-1.5,1.5)
plt.show()
plt.plot([1,1,0,0,-1,0,1,1,-1])
plt.ylim(-1.5,1.5)
plt.xticks(range(9),['02-01','02-02','02-03','02-04',
                     '02-05','02-06','02-07','02-08','02-09'],rotation=45)
plt.show()
```

```python
#2.2 添加文本

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']#中文乱码
plt.title('中国银行2014年收盘价曲线',loc='right')#标题
plt.xlabel('日期') #Y轴标题
plt.ylabel('收盘价')#X轴标题
plt.grid(True,axis='y') #格框

plt.plot(Close['2014'],label='收盘价',color='red',linestyle='solid')#收盘数据
plt.plot(Open['2014'],label='开盘价',color='yellow',linewidth=3)#开盘数据
plt.legend(loc='best')#说明
plt.show()
```

```python
#3常见图形
#3.1柱形
plt.bar([2,3,4,5],height=[228,35,81,1],bottom=2,width=1,color='r',edgecolor='b')
```

```python
#3.2 直方图
plt.hist(Close,range=(2.3,5.5),orientation='vertical',cumulative=True,color='r',edgecolor='b')
'''
bins用于设置直方图分布区间的个数;
range用于设置直方图的小矩形的最小值与最大值;
orientation用于设置直方图的水平或者垂直显示，默认是竖直的直方图，可以将orientation设置为horizontal使其变为水平直方图.
将参数cumulative设置为True,也可以绘制累积分布直方图'''
```

```python
#3.3 饼图
plt.pie([228,35,81,1],labels=('(2,3]','(3,4]',
                             '(4,5]','(5,6]'),colors=('b','g','r','c'),shadow=True)
```

```python
#3.4 箱线图
prcData = ChinaBank.iloc[:,:4]
data = np.array(prcData)
plt.boxplot(data,labels=('Open','High','Low','Close'))
plt.title('中国银行股票箱线图')
'''
矩形的上边和下边分别是变量的上下四分位数，中间的一条线表示数变量的中位数.
在矩形的上下两边各延伸出一条线，每条线的长度一般为1.5倍的四分位距(上下四分位数之差)，
这两条线被视为异常值截断线，上端的线为上边缘线，下端的线为下边缘线，
在线的外面可能还会有一些点，这些点一般会被认为是异常值
'''
```

```python
#4、多图绘制

fig = plt.figure()
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)
ax1.plot(Close['2014'])

ax1.set_xlabel('日期')
ax1.set_ylabel('收盘价')

ax2.plot(Open['2014'])
ax2.set_xlabel('日期')
ax2.set_ylabel('开盘价')

plt.show()
```

```python
Close15 = Close['2015']
fig, ax = plt.subplots()
ax1 = plt.subplot(211)
ax1.plot(Close,color='k')
ax1.set_xlabel('日期')
ax1.set_ylabel('收盘价')
ax1.set_title('中国银行2015年收盘价曲线')
Volume15 = ChinaBank['Volume']['2015']
fig.autofmt_xdate()
Open15 = Open['2015']
ax2 = plt.subplot(212)

left1 = Volume15.index[Close15>=Open15]
hight1 = Volume15[left1]
ax2.bar(left1,hight1,color='r')
left2 = Volume15.index[Close15<Open15]
hight2 = Volume15[left2]
ax2.bar(left2,hight2,color='g')

ax2.set_ylabel('成交量')
ax2.set_title('中国银行2015年成交量柱状图')

fig.tight_layout()
print(left1)
print(left2)
plt.show()
```
